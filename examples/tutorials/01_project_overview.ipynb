## 2. 依赖版本检查

```python
import importlib

def check_package(package_name, min_version=None):
    """检查包是否安装及版本"""
    try:
        pkg = importlib.import_module(package_name)
        version = getattr(pkg, '__version__', 'unknown')
        status = "✅"
        if min_version and version != 'unknown':
            from packaging import version as pkg_version
            if pkg_version.parse(version) < pkg_version.parse(min_version):
                status = "⚠️ 版本过低"
        return status, version
    except ImportError:
        return "❌", "未安装"

print("依赖检查:")
print("-" * 60)

dependencies = [
    ("numpy", "1.19.0"),
    ("pandas", "1.2.0"),
    ("scipy", "1.5.0"),
    ("sklearn", "0.24.0"),
    ("openpyxl", "3.0.0"),
    ("matplotlib", "3.3.0"),
    ("seaborn", "0.11.0"),
]

optional_deps = [
    ("xgboost", "1.4.0"),
    ("lightgbm", "3.2.0"),
    ("catboost", "1.0.0"),
    ("torch", "1.8.0"),
]

print("核心依赖:")
for pkg, min_ver in dependencies:
    status, version = check_package(pkg, min_ver)
    print(f"  {status} {pkg:15s} {version:15s} (需要 >={min_ver})")

print("\n可选依赖:")
for pkg, min_ver in optional_deps:
    status, version = check_package(pkg, min_ver)
    print(f"  {status} {pkg:15s} {version:15s} (可选)")

print("-" * 60)
```

## 5. 基本功能测试

### 5.1 Excel写入功能

```python
import pandas as pd
import numpy as np
from hscredit.report.excel import ExcelWriter

# 创建测试数据
df_test = pd.DataFrame({
    '特征': ['年龄', '收入', '学历'],
    'IV': [0.15, 0.23, 0.08],
    'KS': [0.32, 0.41, 0.25]
})

# 测试Excel写入
writer = ExcelWriter(theme_color='3f1dba')
ws = writer.get_sheet_by_name("Sheet")
writer.insert_df2sheet(ws, df_test, (1, 1), fill=True)

# 保存
output_path = Path("../outputs/api_test_basic.xlsx")
output_path.parent.mkdir(parents=True, exist_ok=True)
writer.save(str(output_path))

print(f"✅ Excel写入测试成功")
print(f"   文件: {output_path}")
print(f"   大小: {output_path.stat().st_size / 1024:.2f} KB")
```

### 5.2 损失函数测试

```python
from hscredit.core.models import FocalLoss, KSMetric

# 测试Focal Loss
y_true = np.array([0, 1, 1, 0, 1])
y_pred = np.array([0.2, 0.8, 0.7, 0.3, 0.9])

focal_loss = FocalLoss(alpha=0.75, gamma=2.0)

# 计算损失
loss_value = focal_loss(y_true, y_pred)
print(f"✅ Focal Loss计算成功")
print(f"   损失值: {loss_value:.4f}")

# 计算梯度
grad = focal_loss.gradient(y_true, y_pred)
print(f"   梯度形状: {grad.shape}")
print(f"   梯度值: {grad}")

# 计算二阶导
hess = focal_loss.hessian(y_true, y_pred)
print(f"   二阶导形状: {hess.shape}")

# 测试KS指标
ks_metric = KSMetric()
ks_value = ks_metric(y_true, y_pred)
print(f"\n✅ KS指标计算成功")
print(f"   KS值: {ks_value:.4f}")
```

### 5.3 框架适配器测试

```python
from hscredit.core.models import LightGBMLossAdapter

# 创建适配器
adapter = LightGBMLossAdapter(focal_loss)

# 获取objective函数
obj_func = adapter.objective()
print(f"✅ LightGBM适配器创建成功")
print(f"   objective函数: {obj_func}")

# 测试objective函数
y_true_lgb = np.array([0, 1, 1, 0, 1])
y_pred_lgb = np.array([0.2, 0.8, 0.7, 0.3, 0.9])

grad_lgb, hess_lgb = obj_func(y_true_lgb, y_pred_lgb)
print(f"   梯度形状: {grad_lgb.shape}")
print(f"   Hessian形状: {hess_lgb.shape}")
```

## 6. Python版本兼容性详细测试

```python
import sys

print(f"Python版本: {sys.version_info.major}.{sys.version_info.minor}.{sys.version_info.micro}")
print("\n兼容性特性测试:")
print("-" * 60)

# 测试1: 类型注解（Python 3.8+）
from typing import List, Dict, Optional, Union, Tuple

def test_func(
    data: List[Dict[str, Union[int, float]]],
    param: Optional[str] = None
) -> Tuple[int, float]:
    return len(data), sum(d.get('value', 0) for d in data)

result = test_func([{'value': 10}, {'value': 20}])
print(f"✅ 类型注解: {result}")

# 测试2: f-string = (Python 3.8+)
x = 10
print(f"✅ f-string调试: {x=}")

# 测试3: 海象运算符 (Python 3.8+)
if (n := len([1, 2, 3, 4, 5])) > 3:
    print(f"✅ 海象运算符: 列表长度{n}大于3")

# 测试4: 位置参数 (Python 3.8+)
def test_positional(a, b, /, c, d, *, e, f):
    return a + b + c + d + e + f

result = test_positional(1, 2, 3, 4, e=5, f=6)
print(f"✅ 位置参数: {result}")

# 测试5: 字典合并运算符 (Python 3.9+)
if sys.version_info >= (3, 9):
    dict1 = {'a': 1, 'b': 2}
    dict2 = {'c': 3, 'd': 4}
    merged = dict1 | dict2
    print(f"✅ 字典合并运算符: {merged}")
    
    # |= 运算符
    dict1 |= dict2
    print(f"✅ 字典更新运算符: {dict1}")
else:
    print("ℹ️  字典合并运算符: 需要Python 3.9+ (当前版本可用基础功能)")

# 测试6: match语句 (Python 3.10+)
if sys.version_info >= (3, 10):
    value = 42
    result = None
    match value:
        case 0:
            result = "零"
        case 42:
            result = "答案"
        case _:
            result = "其他"
    print(f"✅ match语句: {result}")
else:
    print("ℹ️  match语句: 需要Python 3.10+ (当前版本可用基础功能)")

# 测试7: 异常组 (Python 3.11+)
if sys.version_info >= (3, 11):
    try:
        raise ExceptionGroup("多个异常", [
            ValueError("错误1"),
            TypeError("错误2")
        ])
    except ExceptionGroup as e:
        print(f"✅ 异常组: 捕获到{len(e.exceptions)}个异常")
else:
    print("ℹ️  异常组: 需要Python 3.11+ (当前版本可用基础功能)")

print("-" * 60)
print("\n💡 说明:")
print("  ✅ 标记表示当前Python版本支持该特性")
print("  ℹ️  标记表示需要更高Python版本（不影响hscredit使用）")
print("  hscredit核心功能在Python 3.8+所有版本上都能正常工作")
```
