# 分箱表美化展示功能演示

本 Notebook 演示如何在 Jupyter 中使用美化功能展示特征分箱表。

数据来源: hscredit.xlsx（百融数据）

```python
import sys
sys.path.insert(0, '/Users/xiaoxi/CodeBuddy/hscredit/hscredit')

import pandas as pd
import numpy as np
from hscredit.report import feature_bin_stats
from hscredit.utils import style_bin_table
```

```python
# 加载数据
data = pd.read_excel('../utils/hscredit.xlsx')
data['target'] = (data['MOB1'] > 15).astype(int)
print(f"数据形状: {data.shape}")
```

## 1. 默认展示

默认显示：百分比格式 + 进度条高亮 + 简洁风格

```python
# 生成分箱表
table = feature_bin_stats(
    data,
    '百融定制分V9',
    target='target',
    method='optimal',
    max_n_bins=5,
    desc='百融定制分V9'
)

# 美化展示
table.show()
```

## 2. 紧凑模式

只显示核心列

```python
table.show(compact=True)
```

## 3. 关闭百分比格式

```python
table.show(percent_format=False)
```

## 4. 控制高亮选项

```python
# 只显示进度条，不显示IV色阶
table.show(highlight_iv=False)
```

```python
# 只显示坏样本率进度条
table.show(highlight_lift=False, highlight_ks=False)
```

## 5. 多特征对比展示

```python
# 多个评分卡对比
features = ['百融定制分V9', '青云24', '游昆定制分80', '中智小牛分C3']

tables = []
for feat in features:
    table = feature_bin_stats(
        data,
        feat,
        target='target',
        method='optimal',
        max_n_bins=5,
        desc=feat,
        margins=True,
    )
    tables.append(table)

multi_table = pd.concat(tables, axis=0, ignore_index=True)
multi_table.show()
```

## 6. 多目标分析展示

```python
# MOB1 + MOB3 多目标
table = feature_bin_stats(
    data,
    '百融定制分V9',
    overdue=['MOB1', 'MOB2'],
    dpds=15,
    method='optimal',
    max_n_bins=5,
    desc='百融定制分V9'
)
table.show()
```

## 7. 带合计行的展示

```python
# 添加合计行
table = feature_bin_stats(
    data,
    '百融定制分V9',
    target='target',
    method='optimal',
    max_n_bins=5,
    margins=True,
    desc='百融定制分V9'
)
table.show()
```

## 8. 分类变量展示

```python
# 分类变量分析
table = feature_bin_stats(
    data,
    '放款期数',
    target='target',
    desc='放款期数'
)
table.show()
```

## 9. 使用 style_bin_table 函数

```python
# 直接使用 style_bin_table 函数
table = feature_bin_stats(
    data,
    '百融定制分V9',
    target='target',
    method='optimal',
    max_n_bins=5,
    desc='百融定制分V9'
)

# 创建 Styler 对象
styler = style_bin_table(table, percent_format=True, compact=False)
styler
```

## 10. 不同分箱方法对比

```python
methods = ['optimal', 'cart', 'mdlp']

for method in methods:
    print(f"\n=== {method} ===")
    table = feature_bin_stats(
        data,
        '百融定制分V9',
        target='target',
        method=method,
        max_n_bins=5,
        desc=f'百融定制分V9 ({method})'
    )
    table.show(compact=True)
```

## 参数说明

| 参数 | 说明 | 默认值 |
| --- | --- | --- |
| percent_format | 百分比格式显示 | True |
| compact | 紧凑模式，只显示核心列 | False |
| highlight_iv | 高亮IV值（色阶） | True |
| highlight_bad_rate | 高亮坏样本率（进度条） | True |
| highlight_lift | 高亮LIFT值（进度条） | True |
| highlight_ks | 高亮KS值（进度条） | True |
| index_as_bin | 将分箱作为索引 | False |
| high_tech_style | 高科技风格 | False |

## 显示效果说明

- **进度条**: 坏样本率(蓝色)、LIFT值(绿色)、KS值(金色)使用进度条显示
- **色阶**: IV值使用蓝色色阶表示预测能力
- **百分比**: 比例类列默认显示为百分比格式
- **简洁风格**: 浅灰表头、奇偶行交替色、柔和选中色
